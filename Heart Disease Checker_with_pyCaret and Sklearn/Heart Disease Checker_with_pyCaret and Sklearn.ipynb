```python
from pycaret.classification import *
```

```python
setup(df,target='HeartDisease',
     normalize=True,
     numeric_imputation='median',
     
     fix_imbalance=True,
     fix_imbalance_method='randomoversampler',
     log_experiment=True,
     )
```

```python
dum=create_model('dummy')
```

```python
plot_model(dum,plot='confusion_matrix')
```

```python
create_model('knn')
```

```python
tune_model(dum,n_iter=250)
```
